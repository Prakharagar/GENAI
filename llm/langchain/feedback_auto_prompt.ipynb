```python
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
import json
API_KEY=''
```

```python
question = "What are the benefits of using renewable energy?"
answer = "Renewable energy sources such as wind, solar, and hydro power have several benefits. They reduce greenhouse gas emissions, decrease air pollution, and provide a sustainable source of energy."
prompt = "Describe the advantages of renewable energy."
user_feedback = "The prompt could be more specific about the environmental benefits."
context = "Discussion about various energy sources and their impact on the environment."
```

```python
input_prompt={
    "role": "user", 
    "content": f"""Let's improve the quality of prompts by providing detailed corrective feedback 
    based on a thorough analysis of the given inputs. 
        
        Given the following inputs:
        - Query: '{question}'
        - Initial Prompt: '{prompt}'
        - Answer: '{answer}'
        - User Feedback: '{user_feedback}'
        - Context: '{context}'

        Analyze the provided information and evaluate based on the following steps:
        
        1. **Answer Completeness**:
           - Does the answer cover all the aspects asked in the query?
           - Consider user feedback while ranking.
           - Rank from 0 - 5 (0 means completely irrelevant, 5 means most relevant).
           - Provide a brief explanation for your ranking.

        2. **Context Completeness**:
           - Is the context information sufficient to answer the entire query?
           - Rank from 0 - 5 (0 means insufficient, 5 means completely sufficient).
           - Provide an explanation for your ranking.
           - If the rank is 2 or below, it might indicate insufficient information present in the database.

        3. **Context Relevance**:
           - Is the context information relevant to the query?
           - Rank from 0 - 5 (0 means completely irrelevant, 5 means most relevant).
           - Provide an explanation for your ranking.
           - If the rank is 2 or below, it might indicate wrong information got fetched from the database.

        4. **Prompt Effectiveness**:
           - Is the prompt giving the correct direction towards answering the query?
           - Rank from 0 - 5 (0 means completely irrelevant, 5 means most relevant).
           - Provide an explanation for your ranking.
           - If the rank is 2 or below, suggest improvements for the prompt.
           - Suggest the new prompt based on user feedback 

        Please provide your evaluation in the following JSON format:

        {{
            "answer_completeness": {{
                "score": rank1,
                "explanation": explanation
            }},
            "context_completeness": {{
                "score": rank2,
                "explanation": explanation
            }},
            "context_relevance": {{
                "score": rank3,
                "explanation": explanation
            }},
            "prompt": {{
                "score": rank4,
                "explanation": explanation,
                "new_prompt": improved_prompt
            }},
            "labels": ["keyword1", "keyword2", "keyword3"]
        }}
        
        Based on the above evaluation, generate a better prompt that addresses the feedback and provides a clearer direction. The improved prompt should:
        - Be specific and detailed.
        - Emphasize key aspects mentioned in the feedback.
        - Encourage a comprehensive response covering all necessary points.
        - Be clear, engaging, and relevant to the query and context.

        Generate the improved prompt below:
        """   
}
```

```python
client = OpenAI(api_key =API_KEY,)
response = client.chat.completions.create(
                    model="gpt-4-turbo",
                    messages=[input_prompt],
                    temperature=0, 
                    response_format={ "type": "json_object" }
    )
```

```python
def print_dict(d, indent=0):
    
    for key, value in d.items():
        print(' ' * indent + str(key) + ":")
        if isinstance(value, dict):
            print_dict(value, indent + 4)
        else:
            print(' ' * (indent + 4) + str(value))

def main():
    input_json =response.choices[0].message.content
    try:
        llm_output = json.loads(input_json)
        print_dict(llm_output)
        print("######################################################")
        print( 'better_prompt suggestion')
        for key, value in llm_output.items():
            if key=="prompt":
                print(value)
    except json.JSONDecodeError as e:
        print("Error: Invalid JSON format. Please provide valid JSON input.")
        print(e)

if __name__ == "__main__":
    main()
```
